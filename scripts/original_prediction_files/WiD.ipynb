```python
import pandas as pd
import numpy as np
```

```python
reference=pd.read_csv('Sample Reference Sheet.csv')
reference.head()
```

```python
sample_ip=pd.read_csv('sample.csv')
sample_ip.head()
```

```python
sample_ip['prediction']=''
sample_ip.head()
```

```python
for index, r in sample_ip.iterrows():
    
    if (r['External Job Title'] not in reference['Job Title']) and r['Tech']==1:
        [prediction1] = reference[(reference['Job Title']=="Other Technical") & (reference['Gender Rating']==r['Rating'])]['Estimated Percent Female Applicants']
        sample_ip.loc[index ,"prediction"]=prediction1
        
    elif (r['External Job Title'] not in reference['Job Title']) and r['Tech']==0:
        [prediction2] = reference[(reference['Job Title']=="Non-Technical") & (reference['Gender Rating']==r['Rating'])]['Estimated Percent Female Applicants']
        sample_ip.loc[index ,"prediction"]=prediction2
        
    else:
        [prediction3] = reference[(reference['Job Title']== r['External Job Title']) & (reference['Gender Rating']==r['Rating']) & (reference['Tech']==1)]['Estimated Percent Female Applicants']
        sample_ip.loc[index ,"prediction"]=prediction3
        
sample_ip['prediction']
```
